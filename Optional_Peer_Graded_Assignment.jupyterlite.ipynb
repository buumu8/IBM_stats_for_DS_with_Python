### 1. Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)

Null Hypothesis(
): Both average MEDV are the same

Alternative Hypothesis(
): Both average MEDV are NOT the same

```python
boston_df.CHAS.value_counts()
```

```python
a = boston_df[boston_df.CHAS == 0].MEDV
a
```

```python
b = boston_df[boston_df.CHAS == 1].MEDV
b
```

```python
scipy.stats.ttest_ind(a,b,axis=0,equal_var = True)
```

Since p-value more than 0.05, we failed to reject null hypothesis. Thus, MEDV are not the same.

### 2. Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)

```python
boston_df.AGE.value_counts()
```

```python
boston_df
```

the hypothesis

  - H0: mu1 = mu2 = mu3 (the three population means are equal)
  - HA: At least one of the means is different

```python
lo = boston_df[boston_df.age_group == "35 years and younger"].MEDV
mid = boston_df[boston_df.age_group == "between 35 and 70 years"].MEDV
hi = boston_df[boston_df.age_group == "70 years and older"].MEDV
```

```python
f_stats, p_value = scipy.stats.f_oneway(lo,mid,hi,axis = 0)
```

```python
print("F-Statistic={0}, P-value={1}".format(f_stats,p_value))
```

We reject the null hypothesis because P-value is less than 0.05. So at least one of th emeans is different.

### 3. Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)

the hypothesis

  - H0: NOX is not correlated with INDUS
  - HA: NOX is correlated with INDUS

```python
pearson, p_value = scipy.stats.pearsonr(boston_df.NOX,boston_df.INDUS)
```

```python
print("Pearson Coefficient value={0}, P-value={1}".format(pearson,p_value))
```

Since p-value is less than 0.05. we will reject the Null hypothesis and conclude that NOX is correlated with INDUS
